###### Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotnine import *
```

###### Importing library to avoid any exception and warning

```python
import warnings
warnings.filterwarnings('ignore')
```

###### Reading data from the "SampleSuperstore.csv" file

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
df.count()
```

###### Removing the duplicate entries and then watching the dataset

```python
df = df.drop_duplicates()
df.head()
```

###### Dimension of the dataframe after removing the duplicate entries

```python
df.shape
```

###### Remove missing values

```python
df = df.dropna()
df.count()
```

###### Droping the Missing/Null values

```python
print(df.isnull().sum())
```

###### Compute pairwise correlation of columns, excluding NA/null values

```python
df.corr()
```

###### Dropping the column named "Postal Code"

```python
col = ['Postal Code']
df1= df.drop(columns=col,axis=1)
```

###### Two-dimensional, heterogeneous tabular data view

```python
df1
```

###### Compute pairwise correlation of columns, excluding NA/null values

```python
df1.corr()
```

###### To get descriptive statistics

```python
df1.describe()
```

###### Data Visualization using histogram

```python
df1.hist(bins=40,figsize=(18,14))
plt.show()
```

###### Profit and loss data visualization of each "Sub-Category" 

```python
fig_profit = ggplot(df1, aes(x = 'Sub-Category', y='Profit', fill='Sub-Category'))+ geom_col() + coord_flip()+ scale_fill_brewer(type='div', palette='Spectral') + theme_classic()+ ggtitle('Pie Chart')

display(fig_profit)
```

###### Data visualization using heatmap

```python
plt.figure(figsize=(9,6))
sns.heatmap(df1.corr(), cmap="YlGnBu", annot = True)
df1.corr()
```

###### Data visualization using pairplot

```python
plt.figure(figsize=(10,8))
sns.pairplot(df1,hue='Sub-Category', palette="husl",diag_kws=dict(shade=True))
```

###### Data visualization of profits distribution based on the states

```python
plt.figure(figsize=(16,8))
sns.barplot(x=df1['State'].sort_values(ascending=True), y=df1['Profit'])
plt.xticks(rotation=90, fontsize=10)
plt.title('Profits distribution based on the states', fontsize=20)
plt.xlabel('State', fontsize=17)
plt.ylabel('Profit', fontsize=17)
plt.grid(alpha=0.3)
```

###### Distribution of profits based on the different regions i.e  (Central, East,West,South)

```python
C = df1[df1['Region']=='Central']
E = df1[df1['Region']=='East']
W = df1[df1['Region']=='West']
S = df1[df1['Region']=='South']
```

###### Profit distribution based on Central Region

```python
plt.figure(figsize=(12,6))
sns.barplot(x=C['State'].sort_values(ascending=True), y=C['Profit'], saturation=.75)
plt.xticks(rotation=90, fontsize=10);
plt.title('Profit distribution based on Central Region')
plt.xlabel('State',fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(alpha=0.5,linewidth=0.9)
```

###### Profit distribution based on East Region

```python
plt.figure(figsize=(12,6))
sns.barplot(x=E['State'].sort_values(ascending=True), y=E['Profit'], saturation=.75)
plt.xticks(rotation=90, fontsize=10);
plt.title('Profit distribution based on East Region')
plt.xlabel('State',fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(alpha=0.5,linewidth=0.9)
```

###### Profit distribution based on South Region

```python
plt.figure(figsize=(12,6))
sns.barplot(x=S['State'].sort_values(ascending=True), y=S['Profit'], saturation=.75)
plt.xticks(rotation=90, fontsize=10);
plt.title('Profit distribution based on South Region')
plt.xlabel('State',fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(alpha=0.5,linewidth=0.9)
```

###### Profit distribution based on West Region

```python
plt.figure(figsize=(12,6))
sns.barplot(x=W['State'].sort_values(ascending=True), y=W['Profit'], saturation=.75)
plt.xticks(rotation=90, fontsize=10);
plt.title('Profit distribution based on West Region')
plt.xlabel('State',fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(alpha=0.5,linewidth=0.9)
```
