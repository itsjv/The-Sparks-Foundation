###### Importing library

```python
from sklearn.cluster import KMeans
```

###### To get the optimum number of clusters for k-means classification
######  Here We are using Elbow Method to get the number of cluster 

```python
x = iris_data.iloc[:,[0,1,2,3]].values

wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', 
                    max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(7,5))
sns.set(style='whitegrid')    
plt.plot(range(1,11), wcss,'bo--')
plt.title('Elbow method to get Number of Cluster')
plt.xlabel('Number of clusters')
plt.ylabel('Within cluster sum of squares')
plt.show()
```

###### Creating  K-mean Classifier

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(x)
```

###### Visualising Clusters

```python
plt.figure(figsize=(8,7))
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1],
           s=100, c='red', label='Iris-setosa')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1], 
           s=100, c= 'blue', label='Iris-versicolour')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1],
           s=100, c='seagreen', label='Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=100, c='yellow', marker='o', label='Centeroids')
plt.title(" Flowers clusters view ")
plt.legend()
```

###### Conclusion

###### Optimum number of cluster required is 3.
